```python
import gradio as gr
import pandas as pd

from pygwalker.api.gradio import PYGWALKER_ROUTE, get_html_on_gradio

with gr.Blocks() as demo:
    gr.Label("Visual Explore Pricedata in PyGWalker and Gradio")
    gr.Markdown("This is a data app built with pygwalker and gradio library. You can use drag-and-drop operations to explore the data. Start your analysis now!")
    df = pd.read_csv("/Users/svitlanakovalivska/automotive_car_price_estimator/data/cars.csv")
    pyg_html = get_html_on_gradio(df, debug=True)
    gr.HTML(pyg_html)


app = demo.launch(app_kwargs={
    "routes": [PYGWALKER_ROUTE]
})
```
